```python
import skfuzzy as fuzz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
##DATA1.CSV


# Step 1: Read the data
data1 = pd.read_csv('data1.csv')

# Step 2: Normalize or standardize the features
scaler = StandardScaler()
data1_scaled = scaler.fit_transform(data1[['X', 'Y']])

# Step 3: Use FCM algorithm for clustering with c values from 2 to 10
c_values = range(2, 11)
best_fpc = 0
best_c = 0
best_clusters = []

for c in c_values:
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        data1_scaled.T,
        c,
        m=2,
        error=0.005,
        maxiter=1000,
        init=None
    )

    # Assign each data point to the cluster with the highest membership degree
    clusters = [list(u[:, i]).index(max(u[:, i])) for i in range(len(u[0]))]

    # Display 10 Data points with corresponding clusters for the current 'c' value
    print(f"\nMeans-C Fuzzy Clustering with {c} Clusters:")
    print(data1[['X', 'Y', 'Class']].assign(Cluster=clusters).head(10))

    # Visualize the clustering result for the current 'c' value
    plt.scatter(data1['X'], data1['Y'], c=clusters, cmap='viridis')
    plt.title(f'Means-C Fuzzy Clustering with {c} Clusters')
    plt.show()

    # Calculate FPC value
    if fpc > best_fpc:
        best_fpc = fpc
        best_c = c
        best_clusters = clusters

# Step 4: Display the best number of clusters and FPC value
print(f"\nBest number of clusters: {best_c}")
print(f"FPC value: {best_fpc}")

# Step 5: Assign data points to clusters for the best 'c' value
data1['Cluster'] = best_clusters

# Step 6: Display 10 Data points with corresponding clusters for the best 'c' value
print("\n10 Data points with corresponding clusters for the best 'c' value:")
print(data1[['X', 'Y', 'Class', 'Cluster']].head(10))

# # Step 7: Visualize the clustering result for the best 'c' value
# plt.scatter(data1['X'], data1['Y'], c=best_clusters, cmap='viridis')
# plt.title(f'Means-C Fuzzy Clustering with {best_c} Clusters')
# plt.show()
```
